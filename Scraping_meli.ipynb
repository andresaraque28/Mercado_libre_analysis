**En este notebook haremos web scraping a mercado libre usando selenium**

```python
#importamos las librerias
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
```

```python
#abrimos el navegador 
driver = webdriver.Chrome()
```

```python
#ponermos la url a la cual le vamos a hacer el scraping en este caso la de meli para celulares
url='https://listado.mercadolibre.com.co/celular?sb=all_mercadolibre#D[A:celular]'
driver.get(url)
```

```python
# analizamos la paginna y decidimos como sacar la infor y los filtros a usar en este caso:
li_elements = driver.find_elements(By.CSS_SELECTOR, 'section.ui-search-results li') # foltramos para extraer todos los li
data =[] #creamos una lsita vacia para gauerdar datos extraidos
```

```python
# extraer los elementos de la pagina 
for element in li_elements:
    try:
        titulo_element= element.find_element(By.TAG_NAME,'h3')
        precio_element = element.find_element(By.CLASS_NAME,'andes-money-amount__fraction')
        img_element = element.find_element(By.TAG_NAME, 'img')
        imagen_url = (
            img_element.get_attribute('data-src') or # se usa ya que mercado libre tiene una carga dinamica y por tanto no carga todas las imagenes de una vez
            img_element.get_attribute('src') or
            img_element.get_attribute('data-srcset') or
            'No disponible'
        )

        data.append({
            'Titulo': titulo_element.text,
            'Precio': precio_element.text,
            'Imagen': imagen_url
            
        })
    except NoSuchElementException:
        pass
    
```

```python
#ahora guardaremos los datos en dtaframe y posteriormente en un csv
df = pd.DataFrame(data)
df.to_csv('meli_celulare.csv')
```
